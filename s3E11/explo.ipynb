## Machine learning

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression,Ridge,SGDRegressor
from sklearn.preprocessing import RobustScaler
```

```python
X = data.drop("cost",axis=1)
y = data[["cost"]]
```

## Preprocessing data

```python
rs = RobustScaler()
X_proc = rs.fit_transform(X)
```

### Baseline prediction

```python
yhat = [y.mean()[0]]*len(y)
rmse = np.sqrt(mean_squared_error(y,yhat))
print(f"baseline rmse : {round(rmse,2)}")
```

### Linear Regression

```python
lr = LinearRegression()
lr.fit(X_proc,y)
```

```python
yhat = lr.predict(X_proc)
rmse = np.sqrt(mean_squared_error(y,yhat))
print(f"Linear model rmse : {round(rmse,2)}")
```

### Ridge model

```python
ridge = Ridge()
ridge.fit(X_proc,y)
```

```python
yhat = ridge.predict(X_proc)
rmse = np.sqrt(mean_squared_error(y,yhat))
print(f"Ridge model rmse : {round(rmse,2)}")
```

### SGD 

```python
sgd = SGDRegressor(penalty="l1",alpha=1e-3)
sgd.fit(X_proc,y)
```

```python
yhat = sgd.predict(X_proc)
rmse = np.sqrt(mean_squared_error(y,yhat))
print(f"SGD model rmse : {round(rmse,2)}")
```

### Neural network?

```python
from tensorflow.keras import Model,Input,layers
```

```python
inputs = Input(shape=X_proc.shape[1])
x = layers.Dense(32,activation="relu")(inputs)
x = layers.Dense(64,activation="relu")(x)
x = layers.Dense(128,activation="relu")(x)
x = layers.Dense(32,activation="relu")(x)
outputs = layers.Dense(1,activation="linear")(x)

model = Model(inputs=inputs,outputs=outputs)

model.compile(optimizer="adam",loss="mse",metrics="accuracy")
```

```python
model.fit(X_proc,y,epochs=10,validation_split=0.2)
```
